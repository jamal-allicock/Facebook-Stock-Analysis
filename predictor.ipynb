```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
df=pd.read_csv('HistoricalQuotes.csv')
# print(df[' Close/Last'])
for i in range(len(df[' Close/Last'])):
    df[' Close/Last'][i] = float(df[' Close/Last'][i].strip(' $'))
    df[' Open'][i] = float(df[' Open'][i].strip(' $'))
    df[' High'][i] = float(df[' High'][i].strip(' $'))
    df[' Low'][i] = float(df[' Low'][i].strip(' $'))
    # df[' Volume'][i] = df[' Volume'][i].strip(' ')

df.head()
```

```python
plt.figure(figsize=(16,8))
plt.title('Facebook Stock Price')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.plot(df[' Close/Last'])
plt.show()
```

```python
#Number of days to predict into the future
days = 30
df_close = df[[' Close/Last']]
df_close['Prediction']=df_close[[' Close/Last']].shift(-days)
df_close.head
```
